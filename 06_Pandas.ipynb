```python
#Series에 대한 indexing과 slicing
import numpy as np
import pandas as pd

#Series를 생성할 때 index 지정
#여전히 기본 숫자 index는 사용 가능
s = pd.Series([-1, 5, 10, 99], 
             index = ["c","d","a","k"]) #기본 int64
print(s)#index를 다른 것으로 대체한다. 
#내가 정의한 새 인덱스도 동시에 사용할 수 있다. 
#인덱스 값을 키,list를 values의 쌍으로 간주
print(s[0])#그래도 숫자 index는 기본적으로 사용할 수 있다. 
print("s[0]의 값은 : {}".format(s[0]))
print("s[a]의 값은 : {}".format(s['a']))
```

```python
print("s[0:3]의 값은 :\n{}".format(s[0:3])) #slicing
print("s['c':'a']의 값은 :\n{}".format(s["c":"a"])) #s[포함:포함]cf. 숫자는 0부터 시작하므로 하나 뺌
```

```python
#집계 함수 series안의 데이터를 모두 합해서 출력
result = 0.0
for tmp in s: #for 문을 사용하면 느려진다. 지양.
    result += tmp
print(result)
print("Series 안의 데이터 총 합:{}".format(s.sum()) )#value값을 모두 더함. 
```

```python
#Open API를 이용하요 data를 받은 다음.
#해당 data를 처리해서 DataFrame으로 생성해보자. 

import numpy as np
import pandas as pd
import urllib #외부 url호출
import json

day ="20190219"
movie_url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=430156241533f1d058c603178cc3ca0e&targetDt={}".format(day)

movie_page = urllib.request.urlopen(movie_url) #python에서 해당 url 접근, 결과 페이지 받음. 

json_page = json.loads(movie_page.read())

#json_page: dictionary

#박스오피스 순위(index), 제목(column), 당일 매출액(column)

#~~list 생성, for 문. 로직 코드가 들어가야 한다. 

df = pd.DataFrame(json_page) #처음부터 제이손을 데이터 프레임에 넣고, 형태 봄
df
```

```python
json_page["boxOfficeResult"]["dailyBoxOfficeList"][0]["rank"] 
#Dictionary의 안쪽에 있는 값을 알아내는 방법.[0]은 첫번째 데이터 묶음
```

```python
values_rank = list()
values_movieNm = list()
values_salesAmt = list()

for i in json_page["boxOfficeResult"]["dailyBoxOfficeList"]:
    values_rank.append(i["rank"]) #my_dict["name"]#값을 알아내는 방법
    values_movieNm.append(i["movieNm"])
    values_salesAmt.append(i["salesAmt"])
    
new_dict = pd.DataFrame({"movieNm":values_movieNm,"salesAmt":values_salesAmt},
                       index =values_rank )
new_dict
```

```python
#집계함수
#평균(mean): 여러개의 값이 있을 때 그 값들에 대한 합을 개수로 나눈 값
#            수학적 확률의 기대치. 데이터가 많을수록 기대치와 평균이 가까워짐
#편차:(Deviation): 확률변수 X와 평균값에 대한 차이
#             편차의 합은 0으로 수렴. -> 하나의 값으로 도출하기 위해 제곱의 평균을 냄. => 분산
#분산(variance)은 제곱했으므로 단위가 맞지 않음 ->분산에 제곱근을 취하면 표준편차 : 
#분산이 크면 평균으로부터 먼 값이 많고, 분산이 작으면 평균에 값들이 모여있음. 
```

```python
# #공분산
# 독립 사건이면 공분산은 0. 연관성이 없다. 
# 두 개의 확률변수가 연관성이 있으면, 공분산이 연관이 있음. 
# 양의 상관관계, 음의 상관관계 두 가지가 나올 수 있다. 
# 양의 상관관계 -> x 증가시, y 증가. 
# 음의 상관관계 -> x 증가시, y 감소
# 숫자의 절대값 크기로는 판단이 어렵고 양수 음수인지 먼저 본다. 
# 관계가 큰지 적은지는 상관계수로 판단. 
# 상관계수 0 -> 관련성 없음. 
# 절대치가 1에 가까울수록 관계가 큼. 

#공분산 
#확률변수 x와 y의 관계를 알기 위해 사용하는 값
#x가 변할 때 y가 변하는 방향을 알기 위해 사용
#x와 y가 서로 독립사건이면 공분산은 0
#공분산이 양수일 경우 x가 증가하면 y도 증가
#음수일 경우 x가 증가하면 y는 감소하는 경향을 띈다. 
#두 확률변수의 관계 정도 (Strength) 측정에 부적합.
#상관계수로 밀접도 알아낸다. 
```
