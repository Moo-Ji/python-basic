```python
# MovieLens Data Set을 이용( numpy할때 실습한 문제 )
# 1. 사용자가 평가한 모든 영화의 전체의 평균 평점
# 2. 각 사용자별 평균 평점
# 3. 각 영화별 평균 평점
# 4. 가장 평균 평점이 높은 영화의 제목은? (모두 출력)
# 5. comedy영화 중 평점이 가장 낮은 영화의 제목?
# 6. 2015년도에 평가된 모든 Romance 영화의 평균 평점은?
#    timestamp : 1970년 1월 1일 0시 1초 -> 1
# 7. 2014년도에 평가된 Action영화의 평점을 기준으로 
#    상위 10위까지의 영화 제목 출력
```

```python
#모듈 가져오기
import numpy as np
import pandas as pd
```

```python
#CSV파일을 DataFrame으로 읽어오기. 

ratings_df = pd.DataFrame(pd.read_csv("./Data/Movie/ratings.csv"))
display(ratings_df.head())
movies_df = pd.DataFrame(pd.read_csv("./Data/Movie/movies.csv"))
display(movies_df.head())
```

```python
# 1. 사용자가 평가한 모든 영화의 전체의 평균 평점
ratings_df["rating"].mean()
```

```python
# 2. 각 사용자별 평균 평점
u_dict = {}

for i in ratings_df["userId"].unique():
    ID_ratings = pd.DataFrame(ratings_df.loc[ratings_df["userId"] == i]) #값이 i 인 userId
    ID_ratings["rating"].mean()
    u_dict[i] = ID_ratings["rating"].mean()
pd.Series(u_dict)
```

```python
# 3. 각 영화별 평균 평점
m_dict = {}

for i in ratings_df["movieId"].unique():
    MV_ratings = pd.DataFrame(ratings_df.loc[ratings_df["movieId"] == i]) #값이 i 인 userId
    MV_ratings["rating"].mean()
    m_dict[i] = MV_ratings["rating"].mean()
m_df = pd.DataFrame.from_dict( m_dict, orient='index', columns = ["R_mean"])
m_df.index.name = "movieId"
m_df.dropna(how = "any", inplace=True) 
m_df
```

```python
# 4. 가장 평균 평점이 높은 영화의 제목은? (모두 출력)
R_df = pd.merge(movies_df,m_df,left_on="movieId", right_index = True, how ="inner")
R_df.loc[R_df["R_mean"] == R_df["R_mean"].max(),["title","R_mean"]]
```

```python
# 5. comedy영화 중 평점이 가장 낮은 영화의 제목?
Comedy = R_df[R_df["genres"].str.contains("Comedy",na= False)]
Comedy.loc[Comedy["R_mean"]==Comedy["R_mean"].min()]
```

```python
# 6. 2015년도에 평가된 모든 Romance 영화의 평균 평점은?
#    timestamp : 1970년 1월 1일 0시 1초 -> 1
import time
# ts = time.gmtime(964982703)
# time.strftime("%Y",ts)
y_dict = {}
for item in ratings_df["timestamp"]:
    ts = time.gmtime(item)
    result = time.strftime("%Y",ts)
    y_dict[item] = result
    
y_df = pd.DataFrame.from_dict(y_dict, orient = 'index', columns =["year"])
y_df.index.name = "timestamp"
R2_df = pd.merge(ratings_df,y_df,left_on="timestamp", right_index = True, how ="inner")
M_2015 = R2_df.loc[R2_df["year"]=="2015",["movieId","year","rating"]]

Romance = movies_df[movies_df["genres"].str.contains("Romance",na= False)]

R3_df = pd.merge(M_2015,Romance,left_on="movieId", right_index = True, how ="inner")

R3_df["rating"].mean()
```

```python
# 7. 2014년도에 평가된 Action영화의 평점을 기준으로 
#    상위 10위까지의 영화 제목 출력

y_dict = {}
for item in ratings_df["timestamp"]:
    ts = time.gmtime(item)
    result = time.strftime("%Y",ts)
    y_dict[item] = result
    
y_df = pd.DataFrame.from_dict(y_dict, orient = 'index', columns =["year"])
y_df.index.name = "timestamp"
R4_df = pd.merge(ratings_df,y_df,left_on="timestamp", right_index = True, how ="inner")
M_2014 = R2_df.loc[R2_df["year"]=="2014",["movieId","year","rating"]]

Action = movies_df[movies_df["genres"].str.contains("Action",na= False)]
R5_df = pd.merge(M_2014,Action,left_on="movieId", right_index = True, how ="inner")
top_10 = R5_df.sort_values(by=["rating"], ascending=False).copy()
top_10.iloc[0:10][["title","rating","year","genres"]]
```
