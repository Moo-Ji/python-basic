```python
#numpy array연결 (2개의 array를 행, 열 방향으로 연결)
arr1 = np.array([[1,2,3],
                 [4,5,6]]) #2행 3열 2차원 배열
arr2 = np.array([[1,1,1],
                 [2,2,2],
                 [3,3,3],
                 [4,4,4]]) #4행 3열짜리 2차원 배열 
#열 방향으로 붙이려면 행의 개수가 같아야 함. 
# 행 방향으로 붙이려면 열의 개수가 같아야 함. 
result = np.vstack((arr1,arr2)) #수직(vertical) 방향으로 붙인다. (stack)
print(result) #view가 아니고 새로운 배열을 반환. #튜플 형태로 붙이기
```

```python
arr1 = np.array([[1,2,3],
                 [4,5,6]]) #2행 3열 2차원 배열
arr2 = np.array([[9,9,9,9],
                 [9,9,9,9]])#2행 4열짜리 2차원 배열
#두 배열은 행의 개수가 같기 때문에 열 방향으로 붙을 수 있다. 
result = np.hstack((arr1,arr2)) #수평 (horizon)방향으로 붙인다. 
print(result)
```

```python
#numpy broad casting기능 
# 연산을 할 수 있도록 형태를 만든다. 
# 자기 자신을 복사해서 형태를 만들 수 있어야 복사한다. 
arr1 = np.array([[1,2,3],
                 [4,5,6]])

arr2 = np.array([7,8,9]) #-> [[7,8,9],
                            #[7,8.9]]

print(arr1+arr2)
```

```python
#두 개의 배열의 내용이 같은지 전체 비교하는 함수
arr1 = np.arange(10)
arr2 = np.arange(10)

if arr1 is arr2:
    print("같다")
else:
    print("다르다") #객체는 다르다. 

print(np.array_equal(arr1,arr2)) #배열 안의 내용 비교 True
#자바에서는 loop돌려야 한다. 
```

```python
#정렬 관련 함수
#numpy.sort(배열, axis=) : 원본 변화 없고 정렬된 결과 return
#배열객체.sort(axis=) :원본이 정렬됨. return none
np.random.seed(0)
arr = np.random.randint(0,10,(3,3))
print(arr) #원본 출력
arr.sort() #axis를 명시하지 않으면 axis 값은 -1 : none이 아님. 
#axis = -1 (마지막으로 사용할 수 있는 축. 2차원의 경우 axis =1)
print(arr) #원본이 변함. 열 방향으로 오름차순 정렬
arr.sort(axis = 0 ) # 행 방향으로 정렬함. 
print(arr)
```

```python
import numpy as np

ratings = np.loadtxt("./Data/Movie/ratings.csv",
                    delimiter=",", skiprows= 1)
ratings[:5,:] #5행, 전체 열 
#1. 전체 영화에 대한 평균 평점은 얼마인가?
#2. 각 사용자별 평균 평점을 구하시오. 
#3. 각 영화별 평균 평점을 구하시오. 
#4. 가장 평균 평점이 높은 영화의 제목은 무엇인가?
# 동률이 있으면 모두 출력하라. 
R_mean = ratings[:,2] 
print(R_mean.mean())

```

```python
#2. 각 사용자별 평균 평점을 구하시오. 
import numpy as np

for i in np.unique(ratings[:,0]):
    tmp = (ratings[:,0] == i)
    tmp2 = ratings[tmp,2].mean()
    UR_array = np.array([i,tmp2])
    print(UR_array)
```

```python
#3. 각 영화별 평균 평점을 구하시오. 
for i in np.unique(ratings[:,1]):
    tmp = (ratings[:,1] == i)
    tmp2 = ratings[tmp,2].mean()
    MR_array = np.array([i,tmp2])
    print(MR_array)
```

```python
#4. 가장 평균 평점이 높은 영화의 제목은 무엇인가?
# 동률이 있으면 모두 출력하라.
#conda install pandas 설치
import numpy as np
import pandas as pd

movies = pd.read_csv("./data/movie/movies.csv")
movies = movies.values
movies

list=[]
for i in np.unique(ratings[:,1]): #movie id를 각각 뽑아낸다. 
    tmp = (ratings[:,1] == i) # movie id가 같은 값일 때의 참
    tmp2 = ratings[tmp,2].mean() #참값일 때 랭킹의 평균
    MR_array = np.array([i,tmp2]) #movie id와 평균값들을
    list.append(MR_array) #list로 만든다

arr = np.array([0, 0.0]) #초기값을 설정하여 array를 만든다. 
for idx, item in enumerate(list): #list의 인덱스와 값을 튜플로 만든다.
    arr = np.vstack((arr,list[idx])) #초기 값에 list의 index값으로 도출한 결과값을 합

arr = arr[1:, :] #첫 값은 임시값이므로 제하고, 
arr = arr[(arr[:,1] == arr[:,1].max())] #평점이 제일 높은 값들만 참으로 가져옴.
for idx, item in enumerate(arr[:,0]): 
    tmp = movies[(movies[:,0] == item)] #ratings와 movies의 id 값이 같을 때
    print(tmp) #movies의 제목을 뽑아라.   
```
