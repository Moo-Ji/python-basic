```python
import pandas as pd

#1. 파일 불러오기
df=pd.read_csv("Data.csv",sep=",")  
df.head()
```

```python
import pandas as pd

#2. 메뉴별 주문 갯수
df.groupby('Menu').get_group('아메리카노')\
.groupby('Rain').get_group(True)\
.groupby('Date').get_group('2019-02-16')
#groupby('기준').get_group('메뉴명, '비 True/False', '날짜 yyyy-mm-dd')
```

```python
#'' 기준으로 정렬
df.groupby(['Menu','Rain','Date']).first()
```

```python
#메뉴 - 날씨 - 날짜 기준 카운트
df.groupby(['Menu','Rain','Date']).count()
```

```python
#그린티 라떼 갯수 구하기. 비, 날짜별 값 도출 
df.groupby(['Menu','Rain','Date']).count()['Total']['그린티라떼']
```

```python
# 허니브레드를, 비오는 날, 2월 20일에 몇개 먹었나 → 
df.groupby(['Menu','Rain','Date']).count()['Total']['허니브레드'][True]['2019-02-20']
```
