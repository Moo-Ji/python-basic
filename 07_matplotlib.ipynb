```python
%matplotlib nbagg
#nbagg그래프 제어 가능 제어할 필요가 없으면 inline
#numpy를 이용해서  그래프 그리기
#한글 표현이 안된다.
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

#matplotlib 기본 설정
plt.rcParams["axes.grid"] = False #그래프 격자를 표현
plt.rcParams["figure.figsize"] = (8,4) #그래프 크기

#plotting할 데이터를 준비
x_data = range(50)
y_data = np.random.randint(-100,100,50).cumsum()
#cumsum() 앞의 값과의 합을 결과로 도출. 
plt.title("가격변동 추이")
plt.xlabel("기간(일)")
plt.ylabel("가격")

plt.plot(x_data,y_data,"r")
plt.show()
```

```python
%matplotlib nbagg


import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm #한글 나오도록 설정하기. font설정
#matplotlib 기본 설정
plt.rcParams["axes.grid"] = False 
plt.rcParams["figure.figsize"] = (8,4)

#사용할 font의 properties를 생성
font_path = "C:/Windows/Fonts/경기천년바탕_Regular.ttf"
font_prop = fm.FontProperties(fname = font_path, size = 14)
x_data = range(50)
y_data = np.random.randint(-100,100,50).cumsum()

#cumsum() 앞의 값과의 합을 결과로 도출. 
plt.title("가격변동 추이", fontproperties = font_prop)
plt.xlabel("기간(일)", fontproperties = font_prop)
plt.ylabel("가격", fontproperties = font_prop)

plt.plot(x_data,y_data,"r")
plt.show()
```

```python
%matplotlib inline

import numpy as np #데이터 만들기
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

#matplotlib의 기본 설정
plt.rcParams["axes.grid"] = False #True 격자가 보임. 
plt.rcParams["figure.figsize"] = (8,4) #크기
plt.rcParams["axes.unicode_minus"]#범례의 '-'를 글자로 인식해라. 

plt.rcParams["font.size"] =16 #폰트 크기
plt.rcParams["xtick.labelsize"]= 8 
plt.rcParams["ytick.labelsize"]= 8 #y축의 레이블 크기 조정
plt.rcParams["axes.labelsize"] = 14 #label의 크기

#전역으로 사용할 font 설정
font_path = "C:/Windows/Fonts/경기천년바탕_Regular.ttf"
font_family = fm.FontProperties(fname=font_path).get_name()
plt.rcParams["font.family"] = font_family


#plotting할 데이터를 준비
x_data = range(50)
y_data = np.random.randint(-100,100,50).cumsum()
#cumsum() 앞의 값과의 합을 결과로 도출. 
plt.title("가격변동 추이")
plt.xlabel("기간(일)")
plt.ylabel("가격")

plt.plot(x_data,y_data,"r")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

#plotting할 데이터를 준비
s = pd.Series(np.random.randn(10),
             index = np.arange(0,100,10)) #index가 x축
#value를 y축으로 사용
plt.title("가격변동 추이")
plt.xlabel("기간(일)")
plt.ylabel("가격")
plt.plot(s,label="오렌지") 
plt.legend() #범례를 표현
plt.show
```

```python
#DataFrame을 이용한 line plotting
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

#plotting할 데이터를 준비
df = pd.DataFrame(np.random.randint(50,100,(10,3))
                 ,columns = ["이지안","박동훈","홍길동"]
                  , index = np.arange(0,10,1))
#value를 y축으로 사용
plt.title("가격변동 추이")
plt.xlabel("기간(일)")
plt.ylabel("가격")
plt.plot(df) 
plt.show
```

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

#plotting할 데이터를 준비
s = pd.Series(np.random.randint(50,100,(3,)), #50~100사이 숫자 3개
             index = ["국어","영어","수학"])
s.plot(kind ="barh", label = "각 과목 성적" ) #bar - 수직, barh - 수평
# s.plot()과 비슷하지만 plt.plot(s)은 kind 지정 못함
plt.legend()
plt.show
```

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

#plotting할 데이터를 준비
#3명, 4과목 데이터 생성
df1 = pd.DataFrame(np.random.randint(50,100,(3,4))
                 ,columns=["국어", "영어","수학","물리"]
                 ,index = ["이지안","박동훈","홍길동"]) #3명 열, 4행 과목 
#뒤집으면 각 과목당 누가 성적이 좋은지 알 수 있다. 
df2 = pd.DataFrame(np.random.randint(50,100,(4,3))
                 ,index=["국어", "영어","수학","물리"]
                 ,columns = ["이지안","박동훈","홍길동"]) 
df1.plot(kind ="bar" )
df2.plot(kind ="bar" )
plt.legend()
plt.show
```

```python
%matplotlib inline
#산점도(scatter): 두 개의 독립변수 x,y의 관계를 알아보기 위해 사용
#x,y data mapping 우상향 좌하향 등의 분포를 파악하기 쉬움. 

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.random.normal(1,1,(100,1))
y = np.random.normal(-2,4,(100,1))

plt.scatter(x,y)
plt.show()

#랜덤으로 값을 뽑아서 전반적으로 분포 - 데이터 상관관계 없음. 
#상관계수, 공분산 양수 음수 파악 등 가능. 
```
